# Libraries

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
url = 'https://www.basketball-reference.com/players/j/jamesle01.html'
```

# Lebron James Data Viz

```python
lebron_advanced_offense_stats = pd.read_html(url, attrs={"id": "advanced"})[0]
lebron_advanced_offense_stats = lebron_advanced_offense_stats.drop(lebron_advanced_offense_stats.tail(5).index)
lebron_advanced_offense_stats['Age'] = lebron_advanced_offense_stats['Age'].astype(int)
lebron_advanced_stats_offense_filtered = lebron_advanced_offense_stats[['Age', 'OBPM']]
lebron_advanced_stats_offense_filtered
```

```python
plt.scatter(lebron_advanced_stats_offense_filtered['Age'], lebron_advanced_stats_offense_filtered['OBPM'], marker = '.', c = '#FDB927', edgecolor = 'black')
plt.title("Lebron James' Age vs Offensive Box Plus Minus per Season", fontsize = 15)
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Offensive Box Plus Minus', fontsize = 15)
x = lebron_advanced_stats_offense_filtered['Age'].values
s = pd.Series(lebron_advanced_stats_offense_filtered['OBPM'].values, x)
X = sm.add_constant(x)
model = sm.OLS(s, X)
result = model.fit()
y = result.params.loc['x1'] * x + result.params.loc['const']
plt.plot(x, y, c = '#552583')
plt.xticks(lebron_advanced_stats_offense_filtered['Age'].unique())
plt.show()
```

```python
lebron_advanced_defense_stats = pd.read_html(url, attrs={"id": "advanced"})[0]
lebron_advanced_defense_stats = lebron_advanced_defense_stats.drop(lebron_advanced_defense_stats.tail(5).index)
lebron_advanced_defense_stats['Age'] = lebron_advanced_defense_stats['Age'].astype(int)
lebron_advanced_defense_stats_filtered = lebron_advanced_defense_stats[['Age', 'DBPM']]
lebron_advanced_defense_stats_filtered
```

```python
plt.scatter(lebron_advanced_defense_stats_filtered['Age'], lebron_advanced_defense_stats_filtered['DBPM'], marker = '.', c = '#FDB927', edgecolor = 'black')
plt.title("Lebron James' Age vs Defensive Box Plus Minus per Season", fontsize = 15)
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Defensive Box Plus Minus', fontsize = 15)
x = lebron_advanced_defense_stats_filtered['Age'].values
s = pd.Series(lebron_advanced_defense_stats_filtered['DBPM'].values, x)
X = sm.add_constant(x)
model = sm.OLS(s, X)
result = model.fit()
y = result.params.loc['x1'] * x + result.params.loc['const']
plt.plot(x, y, c = '#552583')
plt.xticks(lebron_advanced_defense_stats_filtered['Age'].unique())
plt.show()
```

```python
Lebron_Per_Game = pd.read_html(url, attrs={"id": "per_game"})[0]
Lebron_Per_Game = Lebron_Per_Game.drop(Lebron_Per_Game.tail(5).index)
Lebron_Per_Game['Season'] = Lebron_Per_Game['Season'].astype('string')
Lebron_Per_Game.index = Lebron_Per_Game.index.astype(int)
Lebron_Per_Game_Filtered = Lebron_Per_Game[['Age', 'PTS']]
Lebron_Per_Game_Filtered
```

```python
plt.bar(Lebron_Per_Game_Filtered['Age'], Lebron_Per_Game_Filtered['PTS'], color = '#552583', edgecolor = 'black')
plt.title("Lebron's PPG per Season", fontsize = 15)
plt.xlabel('Season', fontsize = 15)
plt.ylabel('Points per Game', fontsize = 15)
plt.xticks(Lebron_Per_Game_Filtered['Age'].unique())
ax = plt.gca()
ax.set_facecolor('white')
plt.show()
```

```python
lebron_advanced_ws = pd.read_html(url, attrs={"id": "advanced"})[0]
lebron_advanced_ws = lebron_advanced_ws.drop(lebron_advanced_ws.tail(5).index)
lebron_advanced_ws['Age'] = lebron_advanced_ws['Age'].astype(int)
lebron_advanced_ws = lebron_advanced_ws[['Age', 'WS']]
lebron_advanced_ws
```

```python
plt.bar(lebron_advanced_ws['Age'], lebron_advanced_ws['WS'], color = '#552583', edgecolor = 'black')
plt.title("Lebron James' Win Shares vs Age", fontsize = 15)
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Win Shares per Season', fontsize = 15)
plt.xticks(lebron_advanced_ws['Age'].unique())
plt.show()
```
